### Putting it all together

Note: Press 'E'

```python
#Read image
img = cv2.imread('data/messi.jpg',0)

#Display image
cv2.imshow('image',img)

#Shape of the image
image_shape = img.shape
print('image shape: ' + str(image_shape))

#Type of the image
image_type = img.dtype
print('image type: ' + str(image_type))

#Accessing individual pixel value of the image
pixel_value = img[55][75]
print(pixel_value)


k = cv2.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('data/messigray.png',img)
    cv2.destroyAllWindows()
```
